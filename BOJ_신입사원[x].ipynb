# 문제 요약
- 다른 모든 지원자와 비교했을 때 서류심사 성적과 면접시험 성적 중 적어도 하나가 다른 지원자보다 떨어지지 않는 자만 선발
- 지원자 A의 성적이 지원자 B의 성적에 비해 서류 심사 결과와 면접 성적이 모두 떨어진다면 A는 선발되지 않음
- 입사원의 최대 인원수를 구해야함

# Input
- 첫째 줄에는 테스트 케이스의 개수 T(1 ≤ T ≤ 20)가 주어짐
- 각 테스트 케이스의 첫째 줄에 지원자의 숫자 N(1 ≤ N ≤ 100,000)이 주어짐
-  둘째 줄부터 N개 줄에는 각각의 지원자의 서류심사 성적, 면접 성적의 순위가 공백을 사이에 두고 한 줄에 주어짐
- (두 성적 순위는 모두 1위부터 N위까지 동석차 없이 결정된다고 가정)


# 해결과정
- 다시 풀어봐야할 문제 -> 문제 이해 x

# 해설과정
- 여러명의 지원자가 존재
- 서류심사, 면접시험 점수를 고려해서 선발
  - 규칙 : 하위호환인 지원자는 선발될 수 없음
- 각 지원자 마다 누군가의 하위호환인지 확인해야함

<br>

- '서류심사'를 기준으로 정렬해놓음
- 자신보다 위에 배치되어 있는 대상들이 비교대상
- '면접시험' 비교
- 위 과정에서 현재 지원자가 하위호환인지 판단
- 서류 심사를 기준으로 정렬해놓고 밑으로 탐색하면서 비교 대상자 중에 면접 시험 최고 점수와 비교함(주기적으로 갱신) -> O(N)

```python
import sys
input = sys.stdin.readline

T = int(input())

for _ in range(T):
  N = int(input())
  candidates = []

  for _ in range(N):
    s, m = map(int, input().split())
    candidates.append((s, m))

  candidates.sort() # 1차로 s, 2차로 m 으로 정렬

  top_ranking = 1e9 # 무한대
  count = 0

  for i in range(N):
    if candidates[i][1] < top_ranking: # 하위호환이 아닌 경우
      count += 1
      top_ranking = candidates[i][1]

  print(count)
```
